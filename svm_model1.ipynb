# This notebook contains implementation of "SVM"

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
letters_df = pd.read_csv("/content/letterdata.csv")
```

```python
letters_df['letter'].value_counts()
```

```python
letters_df.head(10) 
```

```python
clf = svm.SVC(gamma=0.025, C=10)    
#clf = svm.SVC()
```

```python
X,y = np.array(letters_df)[ :, 1:16], np.array(letters_df.letter)[:] 
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
clf.fit(X_train , y_train)
```

```python
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test, y_pred)
```
